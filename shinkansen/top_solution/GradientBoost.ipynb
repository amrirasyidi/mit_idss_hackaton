```python
def encode_cat_surv(df):
    df['Seat_Class'].iloc[list(df[df['Seat_Class']=='Ordinary']['Seat_Class'].index)]= 0
    df['Seat_Class'].iloc[list(df[df['Seat_Class']=='Green Car']['Seat_Class'].index)]= 1

    for i in list(dfs_train.describe(include=object).T.index):
        df[i].iloc[list(df[df[i]=='Extremely Poor'][i].index)]= 0
        df[i].iloc[list(df[df[i]=='Poor'][i].index)]= 1
        df[i].iloc[list(df[df[i]=='Needs Improvement'][i].index)]= 2
        df[i].iloc[list(df[df[i]=='Acceptable'][i].index)]= 3
        df[i].iloc[list(df[df[i]=='Good'][i].index)]= 4
        df[i].iloc[list(df[df[i]=='Excellent'][i].index)]= 5

    df['Platform_Location'].iloc[list(df[df['Platform_Location']=='Very Inconvenient']['Platform_Location'].index)]= 0
    df['Platform_Location'].iloc[list(df[df['Platform_Location']=='Inconvenient']['Platform_Location'].index)]= 1
    df['Platform_Location'].iloc[list(df[df['Platform_Location']=='Needs Improvement']['Platform_Location'].index)]= 2
    df['Platform_Location'].iloc[list(df[df['Platform_Location']=='Manageable']['Platform_Location'].index)]= 3
    df['Platform_Location'].iloc[list(df[df['Platform_Location']=='Convenient']['Platform_Location'].index)]= 4
    df['Platform_Location'].iloc[list(df[df['Platform_Location']=='Very Convenient']['Platform_Location'].index)]= 5
```

```python
def encode_cat_travel(df):
    df['Gender'] .iloc[list(df[df['Gender']=='Female']['Gender'].index)]= 0
    df['Gender'] .iloc[list(df[df['Gender']=='Male']['Gender'].index)]= 1
    df['Customer_Type'] .iloc[list(df[df['Customer_Type']=='Disloyal Customer']['Customer_Type'].index)]= 0
    df['Customer_Type'] .iloc[list(df[df['Customer_Type']=='Loyal Customer']['Customer_Type'].index)]= 1
    df['Type_Travel'] .iloc[list(df[df['Type_Travel']=='Personal Travel']['Type_Travel'].index)]= 0
    df['Type_Travel'] .iloc[list(df[df['Type_Travel']=='Business Travel']['Type_Travel'].index)]= 1
    df['Travel_Class'] .iloc[list(df[df['Travel_Class']=='Eco']['Travel_Class'].index)]= 0
    df['Travel_Class'] .iloc[list(df[df['Travel_Class']=='Business']['Travel_Class'].index)]= 1
```

```python
for i in [dft_train,dft_test]:
    encode_cat_travel(i)
for j in [dfs_train,dfs_test]:
    encode_cat_surv(j)
```

### Imputing

```python
df_train_copy=df_train.copy()
df_test_copy = df_test.copy()
```

```python
def imput(X):
    li=[]
    for i in range(X.shape[0]):
        for column in list(X.columns):
            if X[column][i] == 'NaN':
                X[column].iloc[i] == np.nan

    for i in range(X.isnull().sum().shape[0]):
        if X.iloc[i].isnull().sum()!=0:
            li.append(list(X.isnull().sum().index)[i])
            
    imputer_mode = SimpleImputer(strategy='most_frequent')
    X[li] = imputer_mode.fit_transform(X[li])
    for i in range(X.isnull().sum().shape[0]):
        if X.iloc[i].isnull().sum()==0:
            li.append(list(X.isnull().sum().index)[i])

    imputer_mode = SimpleImputer(strategy='most_frequent')
    X[li] = imputer_mode.fit_transform(X[li])
    li.append('Departure_Delay_in_Mins')
    li.append('Catering')
    li.append('Onboard_Entertainment')
    li.append('Online_Support')

    return X
```

# Tuning

```python
# Choose the type of classifier
#dtree_estimator = GradientBoost(class_weight = {0: 0.17, 1: 0.83}, random_state = 1)
GBC = GradientBoostingClassifier(learning_rate=0.3,n_estimators=200)

# Grid of parameters to choose from
parameters = {#'max_depth': np.arange(2, 3)
              #'criterion': ['friedman_mse', 'squared_error'],
              #'min_samples_leaf': [1,2,5, 10, 20, 25],
              #'learning_rate': #[.10,.15,.25,.50,.75],
              #'loss' : ['log_loss','exponential'],
              #'n_estimators' : [100,200],
              'min_samples_split':np.arange(2,8)
             }

# Type of scoring used to compare parameter combinations
scorer = metrics.make_scorer(recall_score, pos_label = 1)

# Run the grid search
gridCV = GridSearchCV(GBC, parameters, scoring = scorer, cv = 10)

# Fitting the grid search on the train data
gridCV = gridCV.fit(X_train, y_train)

# Set the classifier to the best combination of parameters
GBC = gridCV.best_estimator_

# Fit the best estimator to the data
GBC.fit(X_train, y_train)
```
